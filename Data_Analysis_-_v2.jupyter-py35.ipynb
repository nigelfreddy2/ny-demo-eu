```python
fields = train_selected.columns.values
selected_features = [i for i in fields if i not in ("Cancel_Ind")]
display(selected_features)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve, auc
import matplotlib.pyplot as plt

model = LogisticRegression()
model.fit(train_selected[selected_features],train_selected[outcome_var])
y_hat = [x[1] for x in model.predict_proba(train_selected[selected_features])]
auc1 = roc_auc_score(train_selected[outcome_var],y_hat)

print("Training AUC: ", auc1)

y_hat2 = [x[1] for x in model.predict_proba(test_selected[selected_features])]
auc2 = roc_auc_score(test_selected[outcome_var],y_hat2)

print("Testing AUC: ", auc2)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
fpr, tpr, threshold = roc_curve(train_selected[outcome_var], y_hat)
roc_auc = auc(fpr, tpr)


plt.title('Receiver Operating Characteristic Train')
plt.plot(fpr, tpr, 'b',
label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.2])
plt.ylim([-0.1,1.2])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
